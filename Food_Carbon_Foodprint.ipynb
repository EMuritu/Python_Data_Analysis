```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= 'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-02-18/food_consumption.csv'
df= pd.read_csv (data)
df.head(35)
```

1. Average CO₂ Emissions per Person by Food Category in East Africa

```python
#East African countries
east_africa = ["Kenya", "Uganda", "Tanzania", "Rwanda", "Ethiopia"]
filtered_df = df[df["country"].isin(east_africa)]
filtered_df.head(60)
```

```python
# Calculate average CO2 emissions per food category
avg_co2 = filtered_df.groupby(['country','food_category'])['co2_emmission'].mean().sort_values(ascending=False)
print(avg_co2)
```

```python
top_emitters = filtered_df.groupby('food_category')['co2_emmission'].mean().nlargest(5)
sns.barplot(x=top_emitters.values, y=top_emitters.index, palette="viridis")
plt.title("Top 5 High-Emission Foods in East Africa")
plt.xlabel("CO₂ Emissions (kg/person/year)")
```

Using data obtained from the Food Carbon Footprint Index 2018, analysis on the average CO2 emissions across five East African countries reveals the following findings:

a. Beef is the highest contributor of CO2 emissions in all five countries (~294.38-78.07 kg CO2/person/year).
b. Dairy products and Lamb follow closely as the top five highest CO2 contributor (~90-135 kg C02/person/year).
c. Plant-based foods (grains, vegetables and legumes) have significantly lower emissions.
c. Ethiopia and Kenya have slightly higher CO2 emissions due to higher meat consumption compared to Rwanda and Uganda.
d. Rwanda has the lowest emissions across most categories.

2. Discrepancy between Consumptions and Emissions: Beef

```python
east_africa = ["Kenya", "Uganda", "Tanzania", "Rwanda", "Ethiopia"]
east_africa_df = df[df["country"].isin(east_africa)]

beef_data = east_africa_df[east_africa_df["food_category"] == "Beef"]

final_data = beef_data[["country", "consumption", "co2_emmission"]]

print("Filtered Beef Data for East Africa:")
final_data.head()
```

While beef is not the most consumed food product in East Africa (compared to grains and vegetables), it has the highest carbon footprint. For instance in Kenya, beef accounts for 294.38 emissions but its consumption rate is merely 9.54. In contrast, wheat and wheat products account for 6.61 emissions while its consumption rate remains 34.65. This highlights how dietary shifts such as reducing beef intake can significantly lower emissions without drastically changing overall food consumption.

3. Animal vs. Non-Animal Products: Impact on Carbon Footprint
Animal based food products (beef, dairy, poulty and fish) contribute to 60-75% of food related CO2 emissions in East Africa, despite a significantly low consumption. Plant-based foods (grains, legumes and vegetables) on the other hand contribute to 25% of emissions with a 70% consumption rate.

4. Beef’s Contribution to CO2 Emissions (Global Perspective)

```python
beef = df[df['food_category'] == 'Beef'].nlargest(20, 'co2_emmission').sort_values('co2_emmission')

plt.figure(figsize=(10, 6))
plt.barh(beef['country'], beef['co2_emmission'], color='#ff6b6b')

plt.title('Top 20 Countries by Beef CO2 Emissions')
plt.xlabel('CO2 Emissions (kg/person/year)')


for i, v in enumerate(beef['co2_emmission']):
    plt.text(v + 10, i, f'{v:.0f}', va='center')

plt.tight_layout()
plt.show()
```

5. Interpretation of CO₂ contributions by food category

```python
df['food_category'] = df['food_category'].replace({
    'Lamb & Goat': 'Lamb & Goat',
    'Milk - inc. cheese': 'Dairy'
})
```

```python
plt.figure(figsize=(12, 6))

category_order = df.groupby('food_category')['co2_emmission'].median().sort_values(ascending=False).index

sns.boxplot(
    data=df,
    x='co2_emmission',
    y='food_category',
    order=category_order,
    palette='viridis', 
    linewidth=1,
    fliersize=3
)

plt.axvline(x=50, color='gray', linestyle='--', alpha=0.5)
plt.text(51, 5, 'High-Impact Threshold', rotation=90, va='center', color='gray')

plt.title("Global CO2 Emissions Distribution by Food Category", 
          fontsize=14, pad=20, fontweight='bold')
plt.xlabel("CO2 Emissions (kg/person/year)")
plt.ylabel("")
plt.grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()
```

6. Top 10 Highest CO2 Emitting Countries

```python
top_10 = (
    df.groupby('country', as_index=False)['co2_emmission']
    .sum()
    .nlargest(10, 'co2_emmission')
    .sort_values('co2_emmission', ascending=True) 
)
```

```python
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

colors = sns.color_palette("flare", len(top_10))

bars = plt.barh(
    top_10['country'],
    top_10['co2_emmission'],
    color=colors,
    edgecolor='black',
    linewidth=0.5
)

for bar in bars:
    width = bar.get_width()
    plt.text(width + 20, bar.get_y() + bar.get_height()/2,
             f'{width:.0f} kg',
             va='center', ha='left',
             fontsize=9)


global_avg = df['co2_emmission'].mean()
plt.axvline(global_avg, color='gray', linestyle='--', alpha=0.7)
plt.text(global_avg + 30, 4.5, f'Global Avg: {global_avg:.0f} kg',
         color='gray', va='center')

plt.title("Top 10 Countries by Total Food-Related CO2 Emissions", 
          fontsize=14, pad=20, fontweight='bold')
plt.xlabel("Total CO2 Emissions (kg/person/year)")
plt.ylabel("")
plt.xlim(0, top_10['co2_emmission'].max() * 1.1)
plt.grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()
```

The Climate Cost of Our Plates: A Data-Driven Exploration of Food Emissions Introduction

Food production accounts for 26% of global greenhouse gas emissions, with significant variations across diets and regions. Using the Food Carbon Footprint Index (2018), this analysis explores how dietary choices drive CO2 emissions worldwide, focusing on:
* East Africa’s emissions profile vs. global trends
* Beef’s disproportionate climate impact
* Policy opportunities for sustainable food systems



The Global Emissions Landscape
Key Finding: Meat and Dairy Dominate Food Emissions. The data reveals a stark hierarchy in emissions intensity:
* Beef (30 kg CO2/kg) and Lamb (25 kg CO2/kg) are the worst offenders.
* Dairy (1.5 kg CO2/kg) and Pork (2 kg CO2/kg) have moderate impacts.
* Plant-based foods (grains, vegetables, nuts) emit <1 kg CO2/kg.

The Top 10 Emitters (kg CO₂/person/year) include:
* Argentina (2,172) – Beef-driven
* Australia (1,939) – Beef/Dairy
* USA (1,719) – Beef/Pork
Kenya, by comparison, emits just 300 kg—80% less than the global average.

Why does this matter?
Just 5 countries (Argentina, Australia, USA, Albania and Newzealand) contribute to 15% of global food emissions despite having only 2% of the population. If these nations cut down their meat consumption by half, global food emissions could possibly drop by 7-10%.

* East Africa’s Contrasting Profile
Key Finding: Low Emissions, but Vulnerable to Dietary Shifts.

While East African nations have low per-capita emissions:
* Kenya: 156 kg CO₂ from beef (5.2 kg consumed/person)
* Ethiopia: 96 kg CO₂ from lamb (2.5 kg consumed/person)

Hidden Risks:
* Rising income could increase meat demand, following Western trends.
* Dairy already accounts for 30% of Kenya’s food emissions due to high consumption (42 kg/person/year).

Opportunity:
* Promoting indigenous plant-based staple foods (millet and legumes) could prevent future emission spikes.

* Beef: The Climate Emergency on Our Plates
Key Finding: Universal 30:1 Emissions Ratio

Whether in Argentina or Kenya, 1 kg beef = 30 kg CO₂ due to:
* Methane from cattle
* Deforestation from grazing
* Feed production and transport

Stark Comparisons:
* Eating a 100g beef burger emits 3 kg CO₂ = driving 8 miles in a gas car.
* Replacing beef with chicken cuts emissions by 90%.

Policy:
* Carbon labeling: Highlight emissions on menus/packaging.
* Subsidies: Shift agricultural incentives to plant-based proteins.

* The Power of Plant-Based Diets
Key Finding: Plants Offer Low-Hanging Fruit

The data demonstrates:
* Grains emit 0.4 kg CO₂/kg
* Vegetables emit 0.2 kg CO₂/kg
* Nuts emit 0.1 kg CO₂/kg

Real-World Impact:
If high-emitting nations replaced 50% of beef with legumes:
* Argentina could cut emissions by 600 kg/person/year.
* Global farmland use could decline by 75%, freeing land for carbon sequestration.

* Policy Recommendations:

A. For High-Emitting Nations:
* Meat taxes: Price beef/lamb to reflect climate costs.
* School meal programs: Prioritize plant-based options.

B. For Low-Emitting Nations (e.g., East Africa):
* Sustainable pathways: Invest in plant-based food security.
* Carbon credits: Reward low-emission farming practices.

C. For Consumers:
* Advocacy for flexible diets: A reduction in meat intake could go a long way in decreasing emissions.
* Local vs. imported: Transportation accounts for <10% of food emissions. The food matters more than where it has been sourced from.

Conclusion: A Crossroads for Food Systems

This analysis reveals two key issues:
* Business as usual as the rising meat consumption will cause +2°C global warming by 2050.
* Sustainable transition: A shift to more sustainable diets could decrease food emissions by 50%.

Ultimately, individual choices do matter, however,  systemic change (policy, industry & education) must be accelerated to enable a sustainable environment for consumption.
