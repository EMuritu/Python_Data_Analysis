### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/09_Time_Series/Apple_Stock/appl_1980_2014.csv)

```python
apple='https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/09_Time_Series/Apple_Stock/appl_1980_2014.csv'
apple_df=pd.read_csv(apple)
apple_df.head(20)
```

### Step 3. Assign it to a variable apple

### Step 4.  Check out the type of the columns

```python
apple_df.dtypes
```

### Step 5. Transform the Date column as a datetime type

```python
apple_df['Date'] = pd.to_datetime(apple_df['Date'])

print(apple_df['Date'].dtype) 
```

### Step 6.  Set the date as the index

```python
apple_df.set_index('Date', inplace=True)
apple_df.head(20)
```

### Step 7.  Is there any duplicate dates?

```python
apple_df.index = pd.to_datetime(apple_df.index)

if apple_df.index.duplicated().any():
    print("Duplicate dates found!")
    print(apple_df[apple_df.index.duplicated(keep=False)])
else:
    print("No duplicate dates in index.")
```

### Step 8.  Ops...it seems the index is from the most recent date. Make the first entry the oldest date.

```python
apple_df = apple_df.sort_index()

print("First date (oldest):", apple_df.index[0])
print("Last date (newest):", apple_df.index[-1])
print("\nFirst 5 rows (oldest first):")
apple_df.head(20)
```

### Step 9. Get the last business day of each month

```python
last_business_days = apple_df.resample('BME').last() 

last_business_days.head(20)
```

### Step 10.  What is the difference in days between the first day and the oldest

```python
oldest_date = apple_df.index[0]   
newest_date = apple_df.index[-1] 


difference_in_days = (newest_date - oldest_date).days
print(f"Number of days between {oldest_date.date()} and {newest_date.date()}: {difference_in_days} days")
```

### Step 11.  How many months in the data we have?

```python
unique_months = apple_df.index.to_period('M').nunique()
print(f"Total unique months in data: {unique_months}")
```

### Step 12. Plot the 'Adj Close' value. Set the size of the figure to 13.5 x 9 inches

```python
ax = apple_df['Adj Close'].plot(
    figsize=(13.5, 9),          
    title='Apple Adjusted Close Price (1980-2014)',
    color='royalblue',
    linewidth=2,
    grid=True
)


ax.set_xlabel('Date')
ax.set_ylabel('Adjusted Close Price (USD)')

import matplotlib.pyplot as plt
plt.show()
```

### BONUS: Create your own question and answer it.

```python
'''Resample the data to quarterly frequency (3 months) using the last trading day of each quarter'''

quarterly = apple_df.resample('QE').agg({  
    'Adj Close': ['first', 'last'],  
    'Volume': 'mean'  
})  
quarterly.columns = ['Open_Price', 'Close_Price', 'Avg_Volume']  

quarterly['Return'] = (quarterly['Close_Price'] - quarterly['Open_Price']) / quarterly['Open_Price'] * 100  

best_quarter = quarterly['Return'].idxmax()  
high_volume_quarter = quarterly['Avg_Volume'].idxmax()  

print(f"Best Quarter: {best_quarter} (Return: {quarterly.loc[best_quarter, 'Return']:.1f}%)")  
print(f"Highest Volume Quarter: {high_volume_quarter} (Volume: {quarterly.loc[high_volume_quarter, 'Avg_Volume']:,.0f} shares)")
```
