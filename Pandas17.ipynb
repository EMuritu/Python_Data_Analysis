### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data)

### Step 3. Assign it to a variable called data and replace the first 3 columns by a proper datetime index.

```python
data='https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data'
data_df = pd.read_csv(data, sep=r"\s+")
data_df["Date"] = pd.to_datetime(data_df[["Yr", "Mo", "Dy"]].astype(str).agg("-".join, axis=1))
data_df = data_df.drop(columns=["Yr", "Mo", "Dy"]).set_index("Date")
data_df.head()
```

```python
data_df.info()
```

### Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.

```python
data_df.index = data_df.index.map(lambda x: x.replace(year=x.year - 100))

data_df.head()
```

### Step 5. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].

```python
data_df.index
data_df.head()
```

### Step 6. Compute how many values are missing for each location over the entire record.  
#### They should be ignored in all calculations below. 

```python
missing_values = data_df.isna().sum()
print(missing_values)
```

```python
missing_pct = data_df.isna().mean() * 100
print(missing_pct.sort_values(ascending=False))
```

### Step 7. Compute how many non-missing values there are in total.

```python
total_non_missing = data_df.count().sum()
print(f"Total non-missing values: {total_non_missing}")
```

```python
data_df.info()
```

### Step 8. Calculate the mean windspeeds of the windspeeds over all the locations and all the times.
#### A single number for the entire dataset.

```python
mean_windspeed = data_df.mean().mean()
print(f"Mean windspeed across all locations and times: {mean_windspeed:.2f}")
```

### Step 9. Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations of the windspeeds at each location over all the days 

#### A different set of numbers for each location.

```python
loc_stats = pd.DataFrame({
    'Min': data_df.min(),      
    'Max': data_df.max(),       
    'Mean': data_df.mean(),     
    'Std Dev': data_df.std()   
})

loc_stats.head(15)
```

### Step 10. Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day.

#### A different set of numbers for each day.

```python
day_stats = pd.DataFrame({
    'Daily Min': data_df.min(axis=1),    
    'Daily Max': data_df.max(axis=1),    
    'Daily Mean': data_df.mean(axis=1),  
    'Daily Std Dev': data_df.std(axis=1) 
})

day_stats.head(15)
```

### Step 11. Find the average windspeed in January for each location.  
#### Treat January 1961 and January 1962 both as January.

```python
january_data = data_df[data_df.index.month == 1]

january_means = january_data.mean().round(2)

print("Average windspeed in January for each location:")
print(january_means)
```

### Step 12. Downsample the record to a yearly frequency for each location.

```python
yearly_data = data_df.resample('YE').mean().round(2)

print("Yearly average windspeeds for each location:")
yearly_data.head()
```

```python
data_df = data_df[(data_df.index.year >= 1960) & (data_df.index.year <= 2000)]
yearly_data = data_df.resample('YE').mean().round(2)

yearly_data.head()
```

### Step 13. Downsample the record to a monthly frequency for each location.

```python
data_df = data_df[(data_df.index.year >= 1960) & (data_df.index.year <= 2000)]
monthly_data = data_df.resample('ME').mean().round(2)

print("Monthly average windspeeds for each location:")
monthly_data.head(12)
```

### Step 14. Downsample the record to a weekly frequency for each location.

```python
data_df = data_df[(data_df.index.year >= 1960) & (data_df.index.year <= 2000)]
weekly_data = data_df.resample('W').mean().round(2)

print("Weekly average windspeeds for each location:")
weekly_data.head(8)
```

### Step 15. Calculate the min, max and mean windspeeds and standard deviations of the windspeeds across all locations for each week (assume that the first week starts on January 2 1961) for the first 52 weeks.

```python

start_date = '1961-01-02'
end_date = pd.to_datetime(start_date) + pd.DateOffset(weeks=52)
weekly_data = data_df.loc[start_date:end_date]

weekly_stats = pd.DataFrame({
    'Week Start': weekly_data.resample('W-MON').first().index,  
    'Min': weekly_data.resample('W-MON').min().min(axis=1),     
    'Max': weekly_data.resample('W-MON').max().max(axis=1),     
    'Mean': weekly_data.resample('W-MON').mean().mean(axis=1),  
    'Std Dev': weekly_data.resample('W-MON').std().mean(axis=1)
}).round(2)

weekly_stats = weekly_stats.reset_index(drop=True)

print("Weekly statistics across all locations for first 52 weeks starting January 2, 1961:")
weekly_stats.head(15)
```
