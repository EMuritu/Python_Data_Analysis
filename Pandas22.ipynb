### Step 1. Import the necessary libraries

### Step 2. Import the dataset from this [address](https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data). 

### Step 3. Assign it to a variable called wine

```python
wine='https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'
wine_df= pd.read_csv(wine)
wine_df.head(20)
```

### Step 4. Delete the first, fourth, seventh, nineth, eleventh, thirteenth and fourteenth columns

```python
columns_to_drop = wine_df.columns[[0, 3, 6, 8, 10, 12, 13]]

wine_df_reduced = wine_df.drop(columns=columns_to_drop)

print("Remaining columns:", wine_df_reduced.columns.tolist())
wine_df_reduced.head()
```

### Step 5. Assign the columns as below:

The attributes are (donated by Riccardo Leardi, riclea '@' anchem.unige.it):  
1) alcohol  
2) malic_acid  
3) alcalinity_of_ash  
4) magnesium  
5) flavanoids

7) hue 

```python
new_column_names = [
    "alcohol", "malic_acid", "alcalinity_of_ash", 
    "magnesium", "flavanoids", "proanthocyanins", "hue"
]

wine_df_reduced.columns = new_column_names

wine_df_reduced.head()
```

### Step 6. Set the values of the first 3 rows from alcohol as NaN

```python
wine_df_reduced.loc[0:2, 'alcohol'] = np.nan

wine_df_reduced.head()
```

### Step 7. Now set the value of the rows 3 and 4 of magnesium as NaN

```python
wine_df_reduced.loc[3:4, 'magnesium'] = np.nan

wine_df_reduced.head()
```

### Step 8. Fill the value of NaN with the number 10 in alcohol and 100 in magnesium

```python
wine_df_reduced['alcohol'] = wine_df_reduced['alcohol'].fillna(10)
wine_df_reduced['magnesium'] = wine_df_reduced['magnesium'].fillna(100)

wine_df_reduced.head()
```

### Step 9. Count the number of missing values

```python
missing_values = wine_df_reduced.isna().sum()

print("Number of missing values per column:")
print(missing_values)
```

### Step 12.  How many missing values do we have?

```python
missing_values = wine_df_reduced.isna().sum()

print("Number of missing values per column:")
print(missing_values)

total_missing = missing_values.sum()
print("\nTotal missing values in the dataset:", total_missing)
```

### Step 13. Delete the rows that contain missing values

```python
wine_df_reduced.dropna(subset=['alcohol'])
```

### Step 14. Print only the non-null values in alcohol

```python
nan_alcohol_rows = wine_df_reduced[wine_df_reduced['alcohol'].isna()]

print("Rows with NaN in 'alcohol':")
print(nan_alcohol_rows[['alcohol']]) 
```

```python
data = {
    'product_id': [101, 102, 103, 104, 105],
    'product_category': ['Electronics', 'Furniture', 'Electronics', np.nan, 'Furniture'],
    'sales': [250, 180, 300, 90, 400],
    'revenue': [5000, np.nan, 6000, 1200, np.nan]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)
```

```python
sales_threshold = df['sales'].quantile(0.75)
high_sales = df[df['sales'] > sales_threshold]
print("\nRows above 75th percentile in sales:\n", high_sales)
```

```python
median_revenue = df['revenue'].median()
df['revenue'] = df['revenue'].fillna(median_revenue)
print("\nDataFrame after filling missing revenue:\n", df)
```

```python
revenue_by_category = df.groupby('product_category', dropna=False)['revenue'].sum()
print("\nTotal revenue per category:\n", revenue_by_category)
```
