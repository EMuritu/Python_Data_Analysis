```python
df_clean=df.dropna(subset=['life_expectancy'])
df_clean.head()
```

```python
df_clean.isnull().sum()
```

```python
missing_columns = df_clean.columns[df_clean.isnull().any()].tolist()
print("Columns with missing values:", missing_columns)
print("\nNumber of missing values per column:")
print(df_clean[missing_columns].isnull().sum())
```

```python
for col in missing_columns:
    if df_clean[col].dtype in ['float64', 'int64']:
        df_clean[col] = df_clean[col].fillna(df_clean[col].median())
    else:
        df_clean[col] = df_clean[col].fillna(df_clean[col].mode()[0])
```

What is the average life expectancy globally?

1. The Average Life Expectancy

```python
average_life_expectancy = df_clean['life_expectancy'].mean()

plt.figure(figsize=(12, 6))
sns.histplot(df_clean['life_expectancy'], bins=20, kde=True, color='skyblue')
plt.axvline(average_life_expectancy, color='red', linestyle='--', linewidth=2, 
            label=f'Average Age: {average_life_expectancy:.1f} years')
plt.title('Distribution of Life Expectancy', fontsize=16)
plt.xlabel('Life Expectancy', fontsize=12)
plt.ylabel('Number of Countries', fontsize=12)
plt.legend()
plt.show()
```

How does life expectancy differ in developing and developed countries?

2. Life Expectancy by Country Status (Developed vs Developing)

```python
#Calculate averages
status_avg = df_clean.groupby('status')['life_expectancy'].mean()
print("\nAverage Life Expectancy by Status:\n", status_avg)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='status', y='life_expectancy', data=df)
plt.title('Life Expectancy Distribution by Country Status')
plt.xlabel('Country Status')
plt.ylabel('Life Expectancy (years)')
plt.show()
```

Developed countries have a significantly higher life expectancy (avg-79 years) compared to developing countries (avg-67 years).Developing countries have a longer lower whisker, indicating more nations with very low life expectancy (e.g., below 50 years). On the other hand, Developed countries have a tight interquartile range (IQR), suggesting consistency in high life expectancy. Some developing countries appear as extreme low outliers.No developed countries fall significantly below their IQR.

2. Top 10 Countries with Highest/Lowest Life Expectancy

```python
top10 = df_clean.groupby('country')['life_expectancy'].mean().nlargest(10)
print("\nTop 10 Countries with Highest Life Expectancy:\n", top10)
```

```python
plt.figure(figsize=(12, 6))
bars = plt.barh(top10.index, top10.values, color='#2c7bb6')

for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.5, bar.get_y() + bar.get_height()/2, 
             f'{width:.1f}', ha='left', va='center')

plt.title('Top 10 Countries with Highest Life Expectancy (2000-2015 Average)', pad=20)
plt.xlabel('Life Expectancy (Years)')
plt.xlim(75, 85)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
bottom10 = df_clean.groupby('country')['life_expectancy'].mean().nsmallest(10)
print("\nTop 10 Countries with Lowest Life Expectancy:\n", bottom10)
```

```python
plt.figure(figsize=(12, 6))
bars = plt.barh(bottom10.index, bottom10.values, color='#d7191c')

for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.5, bar.get_y() + bar.get_height()/2, 
             f'{width:.1f}', ha='left', va='center')

plt.title('Top 10 Countries with Lowest Life Expectancy (2000-2015 Average)', pad=20)
plt.xlabel('Life Expectancy (Years)')
plt.xlim(30, 55)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Top Performers include Japan, Sweden, Iceland and Switzerland while African nations like Sierra Leone, Central African Republic, Lesotho and Angola rank the lowest.

What factors correlate most strongly with life expectancy?

3. Correlation Between Life Expectancy and Key Factors

```python
features = ['life_expectancy', 'adult_mortality', 'bmi', 'gdp', 'schooling', 'alcohol']
corr_matrix = df_clean[features].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix of Key Factors')
plt.show()

# Most positively correlated feature
print("\nMost positively correlated feature:", corr_matrix['life_expectancy'].nlargest(2).index[1])
# Most negatively correlated feature
print("Most negatively correlated feature:", corr_matrix['life_expectancy'].nsmallest(2).index[1])
```

1. Strong Positive Correlations (Close to +1):
- Schooling (0.72): More years of Education translate to longer life expectancy.Education improves health literacy, access to healthcare and socioeconomic status.
- BMI (0.56): Surprisingly positive, BMI's positive correlation contradicts western obesity concerns. In developing nations, low BMI often signals malnutrition while moderate BMI indicates better nutrition.

2. Strong Negative Correlations (Close to -1):
- Adult Mortality (-0.7): Direct inverse relationshipâ€”higher deaths translate to lower life expectancy. There should be a target towards preventable deaths emanating from infections and maternal/child health.

3. Moderate Correlations:
-  GDP (0.43):While wealth matters it is also necessary to focus on equitable access to healthcare and not just economic growth.
-  Alcohol (0.39): This is a weak positive correlation link. It reflects socioeconomic status rather than life expectancy; wealthier nations drink more and have better healthcare.

4.  Life Expectancy Trends Over Time

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='year', y='life_expectancy', hue='status', data=df_clean, errorbar=None)
plt.title('Life Expectancy Trends Over Time (2000-2015)')
plt.xlabel('Year')
plt.ylabel('Average Life Expectancy')
plt.legend(title='Country Status')
plt.show()
```

Life expectancy has been gradually increasing worldwide, with developed countries maintaining a consistent lead.

5. Impact of GDP on Life Expectancy

```python
plt.figure(figsize=(12, 7))
sns.regplot(
    x='gdp', 
    y='life_expectancy', 
    data=df_clean, 
    scatter_kws={'alpha': 0.5, 'color': '#1f77b4'}, 
    line_kws={'color': '#d62728', 'lw': 2},
    logx=True 
)
plt.xscale('log') 
plt.title('GDP vs. Life Expectancy', pad=20)
plt.xlabel('GDP per capita')
plt.ylabel('Life Expectancy (years)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

GDP growth is crucial for developing countries but of more importance is equitable distribution of resources. These resources include education and healthcare. Developed countries with high GDPs have achieved this quite well in their respective countries. GDP growth is also crucial for developing countries but in order to achieve increased life expectancy resources must be equitably distributed.

6. Adult Mortality vs Life Expectancy

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='adult_mortality', y='life_expectancy', data=df_clean, scatter_kws={'alpha':0.3})
plt.title('Adult Mortality vs Life Expectancy')
plt.xlabel('Adult Mortality (per 1000 population)')
plt.ylabel('Life Expectancy')
plt.show()
```

Reducing adult mortality is the single most effective lever for life expectancy gains.Higher deaths translate to lower life expectancy. Increasing life expectancy in developing countries could be achieved through investing in adequate and necesary healthcare infrastructure as well as reducing infectious diseases.

7. Schooling's Impact on Life Expectancy

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='schooling', y='life_expectancy', data=df_clean)
plt.title('Life Expectancy by Schooling Years')
plt.xlabel('Schooling Years')
plt.ylabel('Life Expectancy')
plt.show()
```

Improving education may have compounding benefits.
- Less than 5 years of education translates to a low life expectancy.
- 5-10 years of education brings a steep gain of life expectancy.
- More than 12 years  of education has even better health outcomes and increased life expectancy.
- Developing countries should therefore aim to improve healthcare infrastructure and education.
