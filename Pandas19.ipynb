### Step 6. Print all the Items axis values

### Step 7. Good, now we know  the data avaiable. Create a dataFrame called vol, with the Volume values.

```python
stocks = ["AAPL", "TSLA", "IBM", "META", "GOOGL"]
df = yf.download(stocks, start="2015-01-01", end="2025-07-21", auto_adjust=True)

volume = df['Volume'].copy()  
print(volume.head(10))
```

### Step 8. Aggregate the data of Volume to weekly
#### Hint: Be careful to not sum data from the same week of 2015 and other years.

```python
weekly_volume = df['Volume'].resample('W-MON').sum()

weekly_volume.index = weekly_volume.index.to_series().apply(
    lambda x: f"{x.year}-W{x.isocalendar().week:02d}"
)

print("Top 5 Highest Volume Weeks:")
for ticker in stocks:
    print(f"\n{ticker}:")
    print(weekly_volume[ticker].sort_values(ascending=False).head(5))
```

### Step 9. Find all the volume traded in the year of 2015

```python
volume_2015 = df['Volume']

annual_volume = volume_2015.sum()

print("Total Volume Traded in 2015:")
print(annual_volume)
```

### BONUS: Create your own question and answer it.

calculate and visualize the monthly average trading volume for Apple (AAPL) and Tesla (TSLA) in 2020, highlighting months with unusually high activity (above 1.5x the yearly average)?"

```python
df = yf.download(['AAPL', 'TSLA'], start='2020-01-01', end='2020-12-31', auto_adjust=True)['Volume']

# Calculate monthly averages
monthly_avg = df.resample('ME').mean()

# Calculate yearly average
yearly_avg = df.mean()

# Identify high-activity months (1.5x yearly average)
high_activity = monthly_avg > 1.5 * yearly_avg

print("\nHigh Activity Months (>1.5x Yearly Avg):")
print(monthly_avg[high_activity])
```
